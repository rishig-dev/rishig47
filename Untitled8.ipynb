```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import math
import requests
```

```python
url = "https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt"
text = requests.get(url).text

print(text[:500])
print("\nTotal characters:", len(text))
```

```python
chars = sorted(list(set(text)))
vocab_size = len(chars)

stoi = {ch:i for i,ch in enumerate(chars)}
itos = {i:ch for i,ch in enumerate(chars)}

def encode(s):
    return [stoi[c] for c in s]

def decode(l):
    return ''.join([itos[i] for i in l])

data = torch.tensor(encode(text), dtype=torch.long)
```

```python
block_size = 128
batch_size = 32

def get_batch():
    ix = torch.randint(len(data) - block_size, (batch_size,))
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])
    return x.to(device), y.to(device)
```

```python
class TransformerLM(nn.Module):
    def __init__(self):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, 256)
        self.pos_embedding = nn.Embedding(block_size, 256)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=256,
            nhead=8,
            dim_feedforward=512
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=4)

        self.fc = nn.Linear(256, vocab_size)

    def forward(self, x):
        B, T = x.shape
        token_emb = self.embedding(x)
        pos_emb = self.pos_embedding(torch.arange(T, device=device))
        x = token_emb + pos_emb
        x = self.transformer(x)
        return self.fc(x)
```

```python
model = TransformerLM().to(device)
optimizer = optim.AdamW(model.parameters(), lr=3e-4)
loss_fn = nn.CrossEntropyLoss()
```

```python
epochs = 3000

for step in range(epochs):
    x, y = get_batch()
    logits = model(x)

    loss = loss_fn(logits.view(-1, vocab_size), y.view(-1))

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if step % 300 == 0:
        print(f"Step {step}, Loss: {loss.item():.4f}")
```

```python
def generate(start, length=300):
    model.eval()
    context = torch.tensor(encode(start), dtype=torch.long).unsqueeze(0).to(device)

    for _ in range(length):
        logits = model(context[:, -block_size:])
        probs = torch.softmax(logits[:, -1, :], dim=-1)
        next_char = torch.multinomial(probs, num_samples=1)
        context = torch.cat([context, next_char], dim=1)

    return decode(context[0].tolist())
```

```python
print(generate("KING:", 400))
```

```python
torch.save(model.state_dict(), "transformer_model.pth")
print("Model saved successfully!")
```
